##Desafio: adicionar uma coluna mostrando a região de cada estado

```python
def mapear_regiao(estado):
    estado_str = str(estado)
    first_digit = estado_str[0]

    if first_digit == "1":
        return "Norte"
    elif first_digit == "2":
        return "Nordeste"
    elif first_digit == "3":
        return "Sudeste"
    elif first_digit == "4":
        return "Sul"
    elif first_digit == "5":
        return "Centro-Oeste"
    else:
        return "Total"

df["Região"] = df.reset_index()["Unidade da Federação"].apply(mapear_regiao).tolist()

print(df["Região"])
```

##Desafio: Deixe o gráfico dos top 5 estados que mais gastam mais bonito e detalhado e plote o seu estado de residência

```python
top_5_e_go = df.sort_values("Total", ascending=False).index[1:6].tolist()
top_5_e_go.append("52 Goiás")
top_5_e_go
```

```python
from matplotlib.ticker import NullFormatter

def formatter(x, pos):
    return "R$" + str(round(x / 1e6, 1)) + " milhões"

ax = df.drop(columns=["Total", "Região"]).T.plot(y = top_5_e_go, figsize=(10,6))
plt.legend(bbox_to_anchor=(1,1), labels = ["São Paulo", "Minas Gerais", "Paraná", "Rio Grande do Sul", "Rio de Janeiro", "Goiás"])
plt.ylabel("Valor total processado")
plt.xlabel("Ano/mês")
ax.yaxis.set_major_formatter(formatter)
ax.yaxis.set_minor_formatter(NullFormatter())

ticks2 = [x for x in range(1, 190, 12)]
ax.set_xticks(ticks2)
ax.set_xticklabels([f"200{i}/Jan" for i in range(8,24)])
ax.tick_params(axis='x', which="major", length=5)
plt.xticks(rotation = 45)

plt.title("Top 5 UF's com maior valor total processado de internações e o Goiás")
plt.show()
```
